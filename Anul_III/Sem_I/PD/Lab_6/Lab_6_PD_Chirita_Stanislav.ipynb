```python
#Sarcina 2
import pandas as pd
df_moscov = pd.read_csv('moscov_weather.csv')
df_paris = pd.read_csv('paris_weather.csv')
df_amsterdam = pd.read_csv('amsterdam_weather.csv')
df_moscov['City'] = 'Moscov'
df_paris['City'] = 'Paris'
df_amsterdam['City'] = 'Amsterdam'
df_combined = pd.concat([df_moscov, df_paris, df_amsterdam], ignore_index=True)
df_combined[['H Temp', 'L Temp']] = df_combined[['H Temp', 'L Temp']].apply(pd.to_numeric, errors='coerce')
min_per_day = df_combined.groupby('D')[['H Temp', 'L Temp']].min()
max_per_day = df_combined.groupby('D')[['H Temp', 'L Temp']].max()
mean_per_day = df_combined.groupby('D')[['H Temp', 'L Temp']].mean()
overall_min = df_combined[['H Temp', 'L Temp']].min()
overall_max = df_combined[['H Temp', 'L Temp']].max()
overall_mean = df_combined[['H Temp', 'L Temp']].mean()
print("Valori minime per zi:")
print(min_per_day)
print("\nValori maxime per zi:")
print(max_per_day)
print("\nMedii per zi:")
print(mean_per_day)
print("\nValoare minimă generală:")
print(overall_min)
print("\nValoare maximă generală:")
print(overall_max)
print("\nMedie generală:")
print(overall_mean)
```

```python
#Sarcina 3
import pandas as pd
from datetime import datetime, timedelta
df_combined = pd.read_csv('moscov_weather.csv')
df_combined['D'] = pd.to_datetime(df_combined['D'], format='%d')
df_combined.sort_values(by=['D'], inplace=True)
start_date = df_combined['D'].min()
end_date = df_combined['D'].max()
while start_date <= end_date:
    week_end_date = start_date + timedelta(days=6)
    week_df = df_combined[(df_combined['D'] >= start_date) & (df_combined['D'] <= week_end_date)]
    if not week_df.empty:
        file_name = f"{start_date.strftime('%Y%m%d')}_{week_end_date.strftime('%Y%m%d')}.csv"
        week_df.to_csv(file_name, index=False)
        print(f"Fișierul {file_name} a fost creat.")
    start_date = week_end_date + timedelta(days=1)
```

```python
#Sarcina 4
import pandas as pd
from datetime import datetime
def get_data_for_date(input_date):
    df_combined = pd.read_csv('moscov_weather.csv')
    df_combined['D'] = pd.to_datetime(df_combined['D'], format='%d')
    result_df = df_combined[df_combined['D'].dt.day == input_date.day]
    return result_df if not result_df.empty else None
input_date = datetime(1900, 11, 3)
result = get_data_for_date(input_date)
print(result)
```
